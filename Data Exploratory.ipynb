```python
Range=['Coarse wool price % Change','Copra price % Change','Cotton price % Change','Fine wool price % Change','Hard log price % Change','Hide price % change','Plywood price % Change','Softlog price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

for i in range(len(Range)):
    plt.figure(figsize=(14,9))
    dframe[Range[i]].hist(figsize=(12,9))
    plt.xlabel('Change In Percentage')
    plt.ylabel('Counts')
    plt.legend(Range[i:],loc='upper center',bbox_to_anchor=(1.5,1))
```
